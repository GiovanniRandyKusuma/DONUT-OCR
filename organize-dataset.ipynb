```python
import os
import json
import pandas as pd
import shutil
from tqdm import tqdm
```

```python
# Function to create the folder structure
def create_folder_structure(dataset_name, categories):
    for category in categories:
        os.makedirs(os.path.join(dataset_name, category), exist_ok=True)
        with open(os.path.join(dataset_name, category, "metadata.jsonl"), "w") as metadata_file:
            metadata_file.write("")

# Function to process and write metadata for each image
def write_metadata(file_path, metadata):
    with open(file_path, "a") as metadata_file:
        metadata_file.write(json.dumps(metadata) + "\n")

# Function to extract ground truth parse content
def extract_ground_truth(row):
    # Exclude 'type' and 'file_name' from the row
    row = row.drop(['type', 'file_name'])
    return f'{{"gt_parse": {row.to_json()}}}'

# Function to organize images into folders
def organize_images(dataset_name, categories, df):
    for index, row in tqdm(df.iterrows(), total=len(df), desc="Processing images"):
        print(row)
        print(index)
        category = row['type']
        
        image_path = os.path.join(dataset_name, category, row['file_name'])
        metadata = {
            'file_name': row['file_name'],
            'ground_truth': extract_ground_truth(row)
        }
        shutil.copy('KtpImage/'+row['file_name'], image_path)
        write_metadata(os.path.join(dataset_name, category, "metadata.jsonl"), metadata)


#
```

```python
xlsx_file_path = "KtpImage/ktp_gt.xlsx"
df = pd.read_excel(xlsx_file_path)

# Define dataset name and categories
dataset_name = "ktp_dataset"
categories = ["train", "test", "validation"]

# Create the folder structure
create_folder_structure(dataset_name, categories)

# Organize images and create metadata
organize_images(dataset_name, categories, df)
```
